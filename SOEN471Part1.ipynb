The goal of lemmatization is to reduce a word to its root form, also called a lemma. The function below changes the words in the text to its lemma.

```python
def lemmatization(texts,allowed_postags=["NOUN","ADJ","VERB","ADV"]):
  nlp = spacy.load("en_core_web_sm",disable=["parser","ner"])
  texts_out=[]
  for text in texts:
    doc=nlp(text)
    new_text=[]
    for token in doc:
      if token.pos_ in allowed_postags:
        new_text.append(token.lemma_)
    final = " ".join(new_text)
    texts_out.append(final)
  return texts_out


lemmatized_texts = lemmatization(data[:100])
print(lemmatized_texts[:5])
```

Pass in the lematized text into the gen_words function and it will create a list of a way cleaner text to use and create our LDA topic model.

```python
def gen_words(texts):
  final = []
  for text in texts:
    new = gensim.utils.simple_preprocess(text,deacc=True)
    final.append(new)
  return final

data_words=gen_words(lemmatized_texts)

print(data_words[:5])
```

This now creates a tupple of index of word in dictionary and the actual frequency of that word. Now we have a dictionary for looking up words and their frequency.

```python
id2word = corpora.Dictionary(data_words)

corpus = []
for text in data_words:
  new = id2word.doc2bow(text)
  corpus.append(new)

print(corpus[:50])
```

```python
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,id2word=id2word,num_topics=15,random_state=100,update_every=1,chunksize=100,passes=10,alpha="auto")
```

```python
dictionary = corpora.Dictionary(texts)

# Create a bag-of-words representation of the corpus
corpus = [dictionary.doc2bow(text) for text in texts]

# Train the LDA model
lda_model = gensim.models.LdaModel(corpus, num_topics=15, id2word=dictionary, passes=20)

# Print the topics
for idx, topic in lda_model.print_topics(-1):
    print(f"Topic: {idx + 1}")
    print(topic)
    print()


# for idx, topic in lda_model.print_topics(-1):
#     topic_words = topic.split('+')
#     top_word = topic_words[0].split('*')[1].replace('"', '').strip()
#     print(f"Topic {idx + 1}: {top_word}")
```
